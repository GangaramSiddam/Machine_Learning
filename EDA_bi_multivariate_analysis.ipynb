```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
iris = sns.load_dataset('iris')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue = tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue = tips['smoker'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue = tips['smoker'],style = tips['sex'])
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head(2)
```

```python
sns.barplot(x = titanic['Pclass'],y=titanic['Age'])
```

```python
sns.barplot(x = titanic['Pclass'],y=titanic['Age'],hue = titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = titanic['Sex'] , y = titanic['Age'])
```

```python
sns.boxplot(x = titanic['Sex'] , y = titanic['Age'],hue = titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)


```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist = False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist = False)
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Survived'], titanic['Pclass']))
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

```python
new = flights.groupby('year').sum().reset_index()
```
