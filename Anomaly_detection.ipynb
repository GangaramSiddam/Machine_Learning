```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv('sunday.csv')
```

```python
data.head(5)
```

```python
df = data[['PM1', 'PM2.5', 'PM10', 'Db', 'BPM', 'OX']]
```

```python
df.head()
```

```python
model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1),max_features=1.0)
model.fit(df[['PM1', 'PM2.5', 'PM10', 'Db', 'BPM', 'OX']])
```

```python
df['scores']=model.decision_function(df[['PM1', 'PM2.5', 'PM10', 'Db', 'BPM', 'OX']])
df['anomaly']=model.predict(df[['PM1', 'PM2.5', 'PM10', 'Db', 'BPM', 'OX']])
df.head(20)
```

```python
anomaly=df.loc[df['anomaly']==-1]
anomaly_index=list(anomaly.index)
print(anomaly)
```

```python
def predict_anomaly(model, values):

    df_input = pd.DataFrame([values], columns=['PM1', 'PM2.5', 'PM10', 'Db', 'BPM', 'OX'])
    prediction = model.predict(df_input)
    return prediction[0]



input_values = [38,	63,	68,	61,	89,	98]
predicted_label = predict_anomaly(model, input_values)
print("Predicted Label:", predicted_label)
```
